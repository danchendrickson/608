```python
import numpy as np
import matplotlib.pyplot as plt
from math import exp
```

```python
y=1
t=0
h=0.01
endTime = 2.0
PlotSmoothingFactor = 10.0

Steps = int(endTime / h * PlotSmoothingFactor)
```

```python
lam = -2
```

```python
yexact = []
ycomp=[]
for i in range(Steps+1):
    t = i / Steps * endTime
    yexct=exp(lam*t)
    if i % PlotSmoothingFactor <= 1e-10:
        ycomp.append([t,y])
        y=y+h*(lam*y)
    yexact.append([t,yexct])
```

```python
yexact = np.array(yexact)
ycomp = np.array(ycomp)

plt.figure()
p0=plt.plot(yexact[:,0],yexact[:,1])
p1=plt.plot(ycomp[:,0],ycomp[:,1],'-')
plt.legend(["Exact","Eulers"])
plt.xlabel('t')
plt.ylabel('y')
plt.dpi=1200
plt.figure(figsize=(18, 16)) 
plt.show()
```
