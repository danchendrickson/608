\begin{equation}
\frac{dx}{dt}=f(x,t)
\\
\frac{x_{n+1}-x_{n}}{\Delta t}=f(x_{n},t_{n})
\\
x_{n+1}=x_{n}+\Delta t * f(x_{n},t_{n})
\end{equation}

```python
import numpy as np
import math
import matplotlib.pyplot as plt

dt=0.25
tmax = 10
nmax = int(tmax/dt)

t=[]

t.append(0)
fig = plt.figure()
fig.figsize = [10.0, 7.5]
fig.dpi = 1200
    
def fun(y):
    return 0.5+math.cos(y)

for loops in range(20):
    x=[]
    x.append(loops)
    for n in range(nmax-1):
        xn = x[n]+dt*fun(x[n])
        x.append(xn)
        if loops == 0:
            t.append(t[n]+dt)
    plt.plot(t,x,'.')
```
