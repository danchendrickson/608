```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
alpha1 = 3.0
beta1=0.5
alpha2=0.4
beta2=0.1
```

```python
def f(y,t):
    return np.array([alpha1*y[0]-beta1*y[0]*y[1], \
                    -alpha2*y[1]+beta2*y[0]*y[1]])
```

```python
time = np.linspace(0,70,500)
yinit = np.array([0.1,6])
```

```python
y=odeint(f,yinit,time)
```

```python
plt.figure()
p0=plt.plot(time,y[:,0])
p1=plt.plot(time,y[:,1])
plt.legend(["Prey","Predators"])
plt.xlabel('t')
plt.ylabel('y')
plt.dpi=1200
plt.figure(figsize=(16.0, 9.0)) 
plt.show()
```
